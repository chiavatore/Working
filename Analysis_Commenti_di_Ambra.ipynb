```python
with open('Messagi con Ambra.txt', 'r+',encoding='utf-8') as f:
    contents = f.read()
    print(contents)
```

```python
text = contents
print(text)
```

```python
tokens = word_tokenize(text)
```

```python
print(tokens)
```

```python
print(len(tokens))
```

```python
tokens = [token for token in tokens if token not in '.,:;<>!?[]()`"\'']
```

```python
print(len(tokens))
```

```python
stopwords = stopwords.words("italian")
```

```python
print(stopwords)
```

```python
import nltk
stopwords = nltk.corpus.stopwords.words('italian')
new_words = ('federica','feder', '1', 'ct', '2021', '7', '2', '3', '30' )
for i in new_words:
    stopwords.append(i)
print(stopwords)
```

```python
print(stopwords)
```

```python
tokens = [token for token in tokens if token not in stopwords]
```

```python
print(len(tokens))
print(tokens)
```

```python
stemmer = SnowballStemmer("italian")
```

```python
print(stemmer)
```

```python
print(" ".join(SnowballStemmer.languages))
```

```python
stems = [stemmer.stem(word) for word in tokens]
```

```python
print(len(stems))
print(stems)
```

```python
fdist = FreqDist(stems)
```

```python
print(fdist.most_common(100))
```

```python
df = pd.DataFrame.from_dict(fdist, orient='index')
pd.set_option('display.max_rows', df.shape[0]+1)
df.columns = ['Frequency']
df.index.name = 'Term'
#print(df)
```

```python
sorted_df = df.sort_values(by='Frequency', ascending=False)
sorted_df[sorted_df.Frequency > 3]
print(sorted_df)
```

```python
data_analysis = nltk.FreqDist(stems)

data_analysis.plot(20, cumulative=False)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(stems)
```
