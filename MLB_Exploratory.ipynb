```python
df = df.dropna()
```

```python
# print the first few rows of the DataFrame
print(df.head())
```

```python
# print the DataFrame's shape
print(df.shape)
```

```python
df.rename(columns = {'PITCHER?':'PITCHER'}, inplace = True)
df['PITCHER'] = df['PITCHER'].apply(str)
```

```python
# print the DataFrame's data types
print(df.dtypes)
```

```python
# check for missing values
print(df.isnull().sum())
```

```python
# visualize the distribution of a numeric column
plt.hist(df['age'])
plt.show()
```

```python
# visualize the distribution of a categorical column
df['PITCHER'].value_counts().plot(kind='bar')
plt.show()
```

```python
# calculate basic statistics for a numeric column
print(df['WAR'].describe())
```

```python
# calculate the correlation between two numeric columns
print(df['WAR'].corr(df['age']))
```

```python
# group the data by a categorical column and calculate statistics
grouped_df = df.groupby('PITCHER')['WAR'].mean()
print(grouped_df)
```

```python
# create a scatter plot to visualize the relationship between two numeric columns
plt.scatter(df['salary'], df['WAR'])
plt.xlabel('salary')
plt.ylabel('WAR')
plt.show()
```

```python
# create a box plot to visualize the distribution of a numeric column
plt.boxplot(df['runs_above_rep'])
plt.ylabel('runs_above_rep')
plt.show()
```

```python
# create a bar plot to visualize the mean of a numeric column for each category of a categorical column
df.groupby('PITCHER')['age'].mean().plot(kind='bar')
plt.ylabel('Average Age')
plt.show()
```

```python
# create a pivot table to summarize the data
pivot_table = df.pivot_table(index='team_ID', columns='PITCHER', values='WAR', aggfunc='mean')
print(pivot_table)
```

```python
# create a heatmap to visualize the pivot table
plt.pcolor(pivot_table, cmap='Reds')
plt.colorbar()
plt.show()
```

```python
# create a pairplot to visualize the relationships between multiple numeric columns
import seaborn as sns
sns.pairplot(df, vars=['age', 'WAR', 'salary'])
plt.show()
```

```python
df = df[df["year_ID"] > 1971] 
```

```python
df.sort_values(by=['year_ID'], inplace=True)
```

```python
df = df[df["WAR"] > 6]
```

```python
df.head()
```

```python
df['WAR_G'] = df['WAR']/df['G']
```

```python
df.columns
```

```python
px.bar(df, x='year_ID', y='WAR_G')
```

```python
# Generate base plot:
plot = px.bar(df, x='year_ID', y='WAR', color='WAR',
              color_continuous_scale=px.colors.diverging.RdYlGn)

# Remove colorbar:
plot.update_coloraxes(showscale=False)

# Update plotly style:
plot.update_layout(template='plotly_white')

plot.show()
```

```python
# Label axes:
plot.update_xaxes(title='Year',
                  dtick=1)
plot.update_yaxes(title='Wins Above Replacemen (WAR)')
```

```python
# Update plot layout:
plot.update_layout(
    title=dict(
        text='<b>Wins Above Replacement</b>\
              <br><i><sup>A Visualization of Improvements in Baseball Statistics</sup></i>',
        x=0.085,
        y=0.95,
        font=dict(
            family='Helvetica',
            size=25,
            color='#272b4f'
        )))
```

```python
# Add average MPG across era:
plot.add_hline(y=df['WAR'].mean())

# Add explanation of line:
plot.add_annotation(x=.05,
                    y=0.67,
                    text="Average WAR, Since 1972",
                    textangle=0,
                    xanchor='left',
                    xref="paper",
                    yref="paper",
                    font_color='black',
                    bordercolor='black',
                    borderpad=5,
                    showarrow=True,
                    arrowhead=2,
                    bgcolor='white',
                    arrowside='end'
                    )
```
